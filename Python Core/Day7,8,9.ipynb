```python
import numpy as np

data = np.array([
    [101, 85, 90, 88],
    [102, 78, 82, 80],
    [103, 92, 95, 94],
    [104, 70, 75, 72],
    [105, 88, 91, 89]
])

roll = data[:, 0]
marks = data[:, 1:]

avg = np.mean(marks, axis=1)
topper_index = np.argmax(avg)

print("Average Marks:", avg)
print("Topper Roll No:", roll[topper_index])
```
